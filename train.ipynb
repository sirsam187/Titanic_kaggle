```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt,sklearn as sk,seaborn as sns
```

```python
df=pd.read_csv('/Users/sameersangam/Desktop/Kaggle/Titanic/train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Name','Ticket','Fare','Cabin'],inplace=True,axis=1)
```

```python
dict1={'male':0,'female':'1'}
dict2={'C':0,'S':1,'Q':2}
df.replace({'Sex':dict1,'Embarked':dict2},inplace=True)
df
```

```python
y=df.iloc[:,0].values
X=df.iloc[:,1:].values
X
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(weights='distance',n_neighbors=5)
X=imputer.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
